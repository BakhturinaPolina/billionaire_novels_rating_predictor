```python
# Install and Import Necessary Libraries
# --------------------------------------
!pip install pandas numpy matplotlib seaborn scikit-learn tqdm

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from tqdm import tqdm
import math

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Load the Precomputed Metrics DataFrame
# --------------------------------------
metrics_file_path = '/content/drive/MyDrive/BERTopic_metrics.csv'
df_metrics = pd.read_csv(metrics_file_path)

print("Loaded df_metrics:")
display(df_metrics.head())

print("\nColumns in df_metrics:")
print(df_metrics.columns.tolist())

# Extract 'embedding_model' from 'model_name'
# ------------------------------------------
df_metrics['embedding_model'] = df_metrics['model_name'].apply(
    lambda x: x.split('_')[5] if len(x.split('_')) > 5 else 'Unknown'
)

# Data Cleaning: Removing Runs with Exact 1.0 Values
# ---------------------------------------------------
df_metrics_clean = df_metrics[
    (df_metrics['diversity'] < 1.0) &
    (df_metrics['coherence'] < 1.0)
].copy()

print(f"\nRuns after removing exact 1.0 values: {df_metrics_clean.shape[0]}")

# Data Cleaning: Removing Runs Below 2 Std Below the Mean
# --------------------------------------------------------
coherence_mean = df_metrics_clean['coherence'].mean()
coherence_std = df_metrics_clean['coherence'].std()
topic_diversity_mean = df_metrics_clean['diversity'].mean()
topic_diversity_std = df_metrics_clean['diversity'].std()

coherence_lower = coherence_mean - 2 * coherence_std
topic_diversity_lower = topic_diversity_mean - 2 * topic_diversity_std

print(f"\nCoherence Mean: {coherence_mean:.4f}, Std: {coherence_std:.4f}")
print(f"Topic Diversity Mean: {topic_diversity_mean:.4f}, Std: {topic_diversity_std:.4f}")
print(f"Coherence Lower Bound: {coherence_lower:.4f}")
print(f"Topic Diversity Lower Bound: {topic_diversity_lower:.4f}")

df_metrics_clean = df_metrics_clean[
    (df_metrics_clean['coherence'] >= coherence_lower) &
    (df_metrics_clean['diversity'] >= topic_diversity_lower)
].copy()

print(f"\nRuns after removing outliers below 2 std below the mean: {df_metrics_clean.shape[0]}")

# Visualization: Histograms Before and After Cleaning
# ---------------------------------------------------
plt.figure(figsize=(16, 12))

# Coherence Before Cleaning
plt.subplot(2, 2, 1)
sns.histplot(df_metrics['coherence'], bins=20, kde=True, color='skyblue')
plt.axvline(coherence_mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {coherence_mean:.4f}')
plt.axvline(coherence_mean - coherence_std, color='orange', linestyle='dashed', linewidth=1, label=f'-1 Std: {coherence_mean - coherence_std:.4f}')
plt.axvline(coherence_mean + coherence_std, color='green', linestyle='dashed', linewidth=1, label=f'+1 Std: {coherence_mean + coherence_std:.4f}')
plt.title('Coherence Distribution Before Cleaning')
plt.xlabel('Coherence')
plt.ylabel('Frequency')
plt.legend()

# Coherence After Cleaning
plt.subplot(2, 2, 2)
sns.histplot(df_metrics_clean['coherence'], bins=20, kde=True, color='skyblue')
plt.axvline(df_metrics_clean['coherence'].mean(), color='red', linestyle='dashed', linewidth=1, label=f'Mean: {df_metrics_clean["coherence"].mean():.4f}')
plt.axvline(df_metrics_clean['coherence'].mean() - coherence_std, color='orange', linestyle='dashed', linewidth=1, label=f'-1 Std: {df_metrics_clean["coherence"].mean() - coherence_std:.4f}')
plt.axvline(df_metrics_clean['coherence'].mean() + coherence_std, color='green', linestyle='dashed', linewidth=1, label=f'+1 Std: {df_metrics_clean["coherence"].mean() + coherence_std:.4f}')
plt.title('Coherence Distribution After Cleaning')
plt.xlabel('Coherence')
plt.ylabel('Frequency')
plt.legend()

# Topic Diversity Before Cleaning
plt.subplot(2, 2, 3)
sns.histplot(df_metrics['diversity'], bins=20, kde=True, color='lightgreen')
plt.axvline(topic_diversity_mean, color='red', linestyle='dashed', linewidth=1, label=f'Mean: {topic_diversity_mean:.4f}')
plt.axvline(topic_diversity_mean - topic_diversity_std, color='orange', linestyle='dashed', linewidth=1, label=f'-1 Std: {topic_diversity_mean - topic_diversity_std:.4f}')
plt.axvline(topic_diversity_mean + topic_diversity_std, color='green', linestyle='dashed', linewidth=1, label=f'+1 Std: {topic_diversity_mean + topic_diversity_std:.4f}')
plt.title('Topic Diversity Distribution Before Cleaning')
plt.xlabel('Topic Diversity')
plt.ylabel('Frequency')
plt.legend()

# Topic Diversity After Cleaning
plt.subplot(2, 2, 4)
sns.histplot(df_metrics_clean['diversity'], bins=20, kde=True, color='lightgreen')
plt.axvline(df_metrics_clean['diversity'].mean(), color='red', linestyle='dashed', linewidth=1, label=f'Mean: {df_metrics_clean["diversity"].mean():.4f}')
plt.axvline(df_metrics_clean['diversity'].mean() - topic_diversity_std, color='orange', linestyle='dashed', linewidth=1, label=f'-1 Std: {df_metrics_clean["diversity"].mean() - topic_diversity_std:.4f}')
plt.axvline(df_metrics_clean['diversity'].mean() + topic_diversity_std, color='green', linestyle='dashed', linewidth=1, label=f'+1 Std: {df_metrics_clean["diversity"].mean() + topic_diversity_std:.4f}')
plt.title('Topic Diversity Distribution After Cleaning')
plt.xlabel('Topic Diversity')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()

# Print the Range of Values Above the Mean
# ----------------------------------------
# Coherence: Range above the mean
coherence_above_mean = df_metrics_clean[df_metrics_clean['coherence'] > coherence_mean]['coherence']
if not coherence_above_mean.empty:
    coherence_min_above_mean = coherence_above_mean.min()
    coherence_max = coherence_above_mean.max()
    print(f"\nCoherence range above the mean ({coherence_mean:.4f}): {coherence_min_above_mean:.4f} to {coherence_max:.4f}")
else:
    print(f"\nNo Coherence values above the mean ({coherence_mean:.4f}).")

# Topic Diversity: Range above the mean
diversity_above_mean = df_metrics_clean[df_metrics_clean['diversity'] > topic_diversity_mean]['diversity']
if not diversity_above_mean.empty:
    diversity_min_above_mean = diversity_above_mean.min()
    diversity_max = diversity_above_mean.max()
    print(f"Topic Diversity range above the mean ({topic_diversity_mean:.4f}): {diversity_min_above_mean:.4f} to {diversity_max:.4f}")
else:
    print(f"No Topic Diversity values above the mean ({topic_diversity_mean:.4f}).")

# Reset the Index After Cleaning
# ------------------------------
df_metrics_clean = df_metrics_clean.reset_index(drop=True)

# Print Descriptive Statistics After Cleaning
# -------------------------------------------
print("\nDescriptive Statistics After Cleaning:")
print(df_metrics_clean[['coherence', 'diversity']].describe())

# Proceed with Normalization and Combined Scoring
# -----------------------------------------------
# Initialize the StandardScaler for Z-score normalization
scaler = StandardScaler()

# Apply Z-score normalization to both Coherence and Topic Diversity columns
df_metrics_clean[['Coherence_norm', 'Diversity_norm']] = scaler.fit_transform(df_metrics_clean[['coherence', 'diversity']])

# Define weights for the metrics
weight_coherence_equal = 0.5  # Equal weights
weight_topic_diversity_equal = 0.5

weight_coherence_prioritized = 0.7  # Prioritized coherence
weight_topic_diversity_prioritized = 0.3

# Compute the combined scores
df_metrics_clean['Combined_Score_Equal'] = (
    weight_coherence_equal * df_metrics_clean['Coherence_norm'] +
    weight_topic_diversity_equal * df_metrics_clean['Diversity_norm']
)
df_metrics_clean['Combined_Score_Prioritized'] = (
    weight_coherence_prioritized * df_metrics_clean['Coherence_norm'] +
    weight_topic_diversity_prioritized * df_metrics_clean['Diversity_norm']
)

# Display the first few rows to check combined scores
print("\nCombined Scores:")
display(df_metrics_clean[['Coherence_norm', 'Diversity_norm', 'Combined_Score_Equal', 'Combined_Score_Prioritized']].head())

# Aggregate Coherence and Diversity per Model
# -------------------------------------------
model_aggregated = df_metrics_clean.groupby(['model_name', 'run_index', 'embedding_model']).agg(
    mean_coherence=pd.NamedAgg(column='coherence', aggfunc='mean'),
    mean_diversity=pd.NamedAgg(column='diversity', aggfunc='mean'),
    std_coherence=pd.NamedAgg(column='coherence', aggfunc='std'),
    std_diversity=pd.NamedAgg(column='diversity', aggfunc='std'),
    count_topics=pd.NamedAgg(column='topic_index', aggfunc='count'),
    Coherence_norm=pd.NamedAgg(column='Coherence_norm', aggfunc='mean'),
    Diversity_norm=pd.NamedAgg(column='Diversity_norm', aggfunc='mean'),
    Combined_Score_Equal=pd.NamedAgg(column='Combined_Score_Equal', aggfunc='mean'),
    Combined_Score_Prioritized=pd.NamedAgg(column='Combined_Score_Prioritized', aggfunc='mean')
).reset_index()

print("Aggregated Metrics Per Model:")
display(model_aggregated.head())

# Function to identify Pareto-efficient points
def identify_pareto(df, metrics):
    pareto_efficient = np.ones(df.shape[0], dtype=bool)
    data = df[metrics].values
    for i, row in enumerate(data):
        pareto_efficient[i] = not np.any(np.all((data >= row) & (data > row), axis=1))
    return pareto_efficient

# Identify Pareto-efficient models based on separate metrics
model_aggregated['Pareto_Efficient_Separate'] = identify_pareto(model_aggregated, ['mean_coherence', 'mean_diversity'])

# Select Pareto-efficient models based on separate metrics
pareto_separate = model_aggregated[model_aggregated['Pareto_Efficient_Separate']]

print(f"\nTotal Pareto-efficient models based on separate metrics: {pareto_separate.shape[0]}")

# Identify Pareto-efficient models based on combined scores with equal weights
model_aggregated['Pareto_Efficient_Equal'] = identify_pareto(model_aggregated, ['Coherence_norm', 'Diversity_norm'])

# Select Pareto-efficient models based on combined scores with equal weights
pareto_equal = model_aggregated[model_aggregated['Pareto_Efficient_Equal']]

print(f"\nTotal Pareto-efficient models based on combined scores with equal weights: {pareto_equal.shape[0]}")

# Identify Pareto-efficient models based on combined scores with prioritized coherence
model_aggregated['Pareto_Efficient_Prioritized'] = identify_pareto(model_aggregated, ['Coherence_norm', 'Diversity_norm'])

# Select Pareto-efficient models based on combined scores with prioritized coherence
pareto_prioritized = model_aggregated[model_aggregated['Pareto_Efficient_Prioritized']]

print(f"\nTotal Pareto-efficient models based on combined scores with prioritized coherence: {pareto_prioritized.shape[0]}")

# Display Pareto-efficient models
print("\nPareto-efficient Models Based on Separate Metrics:")
display(pareto_separate[['model_name', 'run_index', 'embedding_model', 'mean_coherence', 'mean_diversity']])

print("\nPareto-efficient Models Based on Combined Scores with Equal Weights:")
display(pareto_equal[['model_name', 'run_index', 'embedding_model', 'Combined_Score_Equal']])

print("\nPareto-efficient Models Based on Combined Scores with Prioritized Coherence:")
display(pareto_prioritized[['model_name', 'run_index', 'embedding_model', 'Combined_Score_Prioritized']])

# Identify Pareto-efficient topics within each model
def identify_pareto_group(group, metrics):
    data = group[metrics].values
    pareto_efficient = np.ones(data.shape[0], dtype=bool)
    for i, row in enumerate(data):
        pareto_efficient[i] = not np.any(np.all((data >= row) & (data > row), axis=1))
    return pd.Series(pareto_efficient, index=group.index)

# Apply Pareto efficiency to topics within each group
df_metrics_clean['Pareto_Efficient_Topic'] = df_metrics_clean.groupby(['model_name', 'run_index', 'embedding_model'], group_keys=False).apply(
    lambda group: identify_pareto_group(group, ['coherence', 'diversity'])
)

print("\nSample of Pareto-efficient topics within each model:")
display(df_metrics_clean[['model_name', 'run_index', 'embedding_model', 'topic_index', 'coherence', 'diversity', 'Pareto_Efficient_Topic']].head(10))

# Function to calculate percentage of Pareto-efficient topics per model
def calculate_pareto_percentage(df, group_cols, score_type='separate'):
    if score_type == 'separate':
        # For separate metrics, count topics Pareto-efficient based on separate coherence and diversity
        df['Pareto_Separate'] = df['Pareto_Efficient_Topic'] & \
                                 (df['coherence'] >= coherence_mean) & \
                                 (df['diversity'] >= topic_diversity_mean)
        summary = df.groupby(group_cols).agg(
            total_topics=pd.NamedAgg(column='topic_index', aggfunc='count'),
            pareto_topics=pd.NamedAgg(column='Pareto_Separate', aggfunc='sum')
        ).reset_index()
        summary['Pareto_Percentage_Separate'] = (summary['pareto_topics'] / summary['total_topics']) * 100
        return summary[['model_name', 'run_index', 'embedding_model', 'Pareto_Percentage_Separate']]

    elif score_type == 'equal':
        # For combined equal weights, compute Pareto-efficient based on combined scores
        df['Combined_Score_Equal'] = (0.5 * df['Coherence_norm']) + (0.5 * df['Diversity_norm'])
        pareto_flags = df.groupby(group_cols, group_keys=False).apply(
            lambda group: identify_pareto_group(group, ['Combined_Score_Equal', 'Combined_Score_Equal'])
        )
        df['Pareto_Efficient_Combined_Equal'] = pareto_flags
        # Ensure no NaN values
        df['Pareto_Efficient_Combined_Equal'] = df['Pareto_Efficient_Combined_Equal'].fillna(False)
        summary = df.groupby(group_cols).agg(
            total_topics=pd.NamedAgg(column='topic_index', aggfunc='count'),
            pareto_topics=pd.NamedAgg(column='Pareto_Efficient_Combined_Equal', aggfunc='sum')
        ).reset_index()
        summary['Pareto_Percentage_Equal'] = (summary['pareto_topics'] / summary['total_topics']) * 100
        return summary[['model_name', 'run_index', 'embedding_model', 'Pareto_Percentage_Equal']]

    elif score_type == 'prioritized':
        # For combined prioritized coherence
        df['Combined_Score_Prioritized'] = (0.7 * df['Coherence_norm']) + (0.3 * df['Diversity_norm'])
        pareto_flags = df.groupby(group_cols, group_keys=False).apply(
            lambda group: identify_pareto_group(group, ['Combined_Score_Prioritized', 'Combined_Score_Prioritized'])
        )
        df['Pareto_Efficient_Combined_Prioritized'] = pareto_flags
        # Ensure no NaN values
        df['Pareto_Efficient_Combined_Prioritized'] = df['Pareto_Efficient_Combined_Prioritized'].fillna(False)
        summary = df.groupby(group_cols).agg(
            total_topics=pd.NamedAgg(column='topic_index', aggfunc='count'),
            pareto_topics=pd.NamedAgg(column='Pareto_Efficient_Combined_Prioritized', aggfunc='sum')
        ).reset_index()
        summary['Pareto_Percentage_Prioritized'] = (summary['pareto_topics'] / summary['total_topics']) * 100
        return summary[['model_name', 'run_index', 'embedding_model', 'Pareto_Percentage_Prioritized']]

    else:
        raise ValueError("Invalid score_type. Choose from 'separate', 'equal', 'prioritized'.")

# Calculate Pareto-efficient topic percentages per model for different scoring schemes

# a. Separate Metrics
pareto_percentage_separate = calculate_pareto_percentage(
    df_metrics_clean,
    ['model_name', 'run_index', 'embedding_model'],
    score_type='separate'
)

# b. Combined Scores with Equal Weights
pareto_percentage_equal = calculate_pareto_percentage(
    df_metrics_clean,
    ['model_name', 'run_index', 'embedding_model'],
    score_type='equal'
)

# c. Combined Scores with Prioritized Coherence
pareto_percentage_prioritized = calculate_pareto_percentage(
    df_metrics_clean,
    ['model_name', 'run_index', 'embedding_model'],
    score_type='prioritized'
)

# Merge all percentages into a single DataFrame
pareto_percentage = pareto_percentage_separate.merge(
    pareto_percentage_equal,
    on=['model_name', 'run_index', 'embedding_model'],
    how='left'
).merge(
    pareto_percentage_prioritized,
    on=['model_name', 'run_index', 'embedding_model'],
    how='left'
)

print("\nPercentage of Pareto-efficient Topics Per Model:")
display(pareto_percentage.head())

# Best Performing Models Based on Aggregated Metrics

# a. Separate Metrics
best_separate = pareto_separate.sort_values(
    ['mean_coherence', 'mean_diversity'],
    ascending=[False, False]
).head(10)
print(f"\nTop 10 Best Performing Models Based on Separate Metrics:")
display(best_separate[['model_name', 'run_index', 'embedding_model', 'mean_coherence', 'mean_diversity']])

# b. Combined Scores with Equal Weights
best_equal = pareto_equal.sort_values(
    'Combined_Score_Equal',
    ascending=False
).head(10)
print(f"\nTop 10 Best Performing Models Based on Combined Scores with Equal Weights:")
display(best_equal[['model_name', 'run_index', 'embedding_model', 'Combined_Score_Equal']])

# c. Combined Scores with Prioritized Coherence
best_prioritized = pareto_prioritized.sort_values(
    'Combined_Score_Prioritized',
    ascending=False
).head(10)
print(f"\nTop 10 Best Performing Models Based on Combined Scores with Prioritized Coherence:")
display(best_prioritized[['model_name', 'run_index', 'embedding_model', 'Combined_Score_Prioritized']])

# Best Performing Models Based on Percentage of Pareto-efficient Topics

# a. Separate Metrics
best_pareto_separate = pareto_percentage.sort_values(
    'Pareto_Percentage_Separate',
    ascending=False
).head(10)
print(f"\nTop 10 Best Performing Models Based on Percentage of Pareto-efficient Topics (Separate Metrics):")
display(best_pareto_separate)

# b. Combined Scores with Equal Weights
best_pareto_equal = pareto_percentage.sort_values(
    'Pareto_Percentage_Equal',
    ascending=False
).head(10)
print(f"\nTop 10 Best Performing Models Based on Percentage of Pareto-efficient Topics (Combined Equal Weights):")
display(best_pareto_equal)

# c. Combined Scores with Prioritized Coherence
best_pareto_prioritized = pareto_percentage.sort_values(
    'Pareto_Percentage_Prioritized',
    ascending=False
).head(10)
print(f"\nTop 10 Best Performing Models Based on Percentage of Pareto-efficient Topics (Combined Prioritized Coherence):")
display(best_pareto_prioritized)

# Save the top models to Google Drive
best_separate.to_csv('/content/drive/MyDrive/top_best_separate_metrics.csv', index=False)
best_equal.to_csv('/content/drive/MyDrive/top_best_combined_equal_weights.csv', index=False)
best_prioritized.to_csv('/content/drive/MyDrive/top_best_combined_prioritized_coherence.csv', index=False)
best_pareto_separate.to_csv('/content/drive/MyDrive/top_best_pareto_percentage_separate.csv', index=False)
best_pareto_equal.to_csv('/content/drive/MyDrive/top_best_pareto_percentage_equal.csv', index=False)
best_pareto_prioritized.to_csv('/content/drive/MyDrive/top_best_pareto_percentage_prioritized.csv', index=False)

print("\nTop performing models have been successfully saved to your Google Drive.")

# Create an Additional DataFrame Gathering All Pareto-efficient Models
# --------------------------------------------------------------------

# Merge Pareto-efficient models data
all_pareto_models = pd.merge(
    pareto_separate[['model_name', 'run_index', 'embedding_model', 'mean_coherence', 'mean_diversity']],
    pareto_equal[['model_name', 'run_index', 'embedding_model', 'Combined_Score_Equal']],
    on=['model_name', 'run_index', 'embedding_model'],
    how='outer'
).merge(
    pareto_prioritized[['model_name', 'run_index', 'embedding_model', 'Combined_Score_Prioritized']],
    on=['model_name', 'run_index', 'embedding_model'],
    how='outer'
).merge(
    pareto_percentage,
    on=['model_name', 'run_index', 'embedding_model'],
    how='left'
)

# Add a column indicating which Pareto-efficient criteria each model meets
all_pareto_models['Pareto_Efficient_Criteria'] = all_pareto_models.apply(
    lambda row: ', '.join(filter(None, [
        'Separate Metrics' if not pd.isnull(row['mean_coherence']) and not pd.isnull(row['mean_diversity']) else None,
        'Equal Weights' if not pd.isnull(row['Combined_Score_Equal']) else None,
        'Prioritized Coherence' if not pd.isnull(row['Combined_Score_Prioritized']) else None
    ])),
    axis=1
)

# Rearrange columns
columns_order = [
    'model_name', 'run_index', 'embedding_model', 'Pareto_Efficient_Criteria',
    'mean_coherence', 'mean_diversity', 'Combined_Score_Equal', 'Combined_Score_Prioritized',
    'Pareto_Percentage_Separate', 'Pareto_Percentage_Equal', 'Pareto_Percentage_Prioritized'
]
all_pareto_models = all_pareto_models[columns_order]

# Sort the DataFrame based on Combined Scores and Percentages
all_pareto_models.sort_values(
    by=['Combined_Score_Prioritized', 'Combined_Score_Equal', 'mean_coherence', 'mean_diversity'],
    ascending=False,
    inplace=True
)

print("\nAll Pareto-efficient Models with Combined Scores and Percentage of Best Performing Topics:")
display(all_pareto_models.head(10))

# Save this DataFrame to Google Drive
all_pareto_models.to_csv('/content/drive/MyDrive/all_pareto_efficient_models.csv', index=False)
print("\nAll Pareto-efficient models have been saved to your Google Drive as 'all_pareto_efficient_models.csv'.")

```
